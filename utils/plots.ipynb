```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import colormaps
```

```python
def calc_action(pos, target):
    """
    Calculate optimal piezo to actuate
    :param pos:    position
    :param offset:  Offset to target in pixel difference np.array([x, y])
    :return:        integer in [1, 2, 3, 4] with len() = number of piezos
    """
    angle = np.arctan2(target[1] - pos[1], target[0] - pos[0])
    angle = np.degrees(angle)
    # return angle
    angle_partition = (angle + 382.5) % 360 // 45
    return angle_partition
```

```python
def angle_part(angle):
    return (angle + 382.5) % 360 // 45
```

```python
angles = np.linspace(-360, 360, 500)
angle_partition = (angles + 382.5) % 360 // 45
# plt.plot(angles, angle_partition)
cmap = plt.get_cmap('Paired')
for i in angles:
    plt.polar(np.radians(i), 1, marker='o', linestyle='None', label=f'Angle Partition, {angle_part(i)}', color=cmap(angle_part(i)/8))
# plt.polar(np.radians(angles), angle_partition, marker='o', linestyle='None', label='Angle Partition')
plt.title('Angle Partition based on Angle')
# plt.legend()
plt.show()

```

```python
points = np.linspace(-np.pi/4, np.pi, 200)
points_x = np.cos(points)
points_y = np.sin(points)
angles = []
# cmap = plt.get_cmap('viridis')
for x, y in zip(points_x, points_y):
    print(f"{x:.2f},{y:.2f}", calc_action((0,0), (x, y)))
    angles.append(calc_action((0,0), (x, y)))

plt.scatter(points_x, points_y, c=angles, cmap='Paired', label=angles)
# plt.legend()
 
```
